```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from PIL import ImageFilter
```

```python
path = "Anya.jpg"
```

```python
img = Image.open(path)
display(img)
```

```python
img_np = np.asarray(img)
img_np.shape
```

```python
plt.imshow(img_np[50:, :, :])             # crop upper part
img_50 = Image.fromarray(img_np[50:, :, :])
```

```python
img_blur = img_50.filter(ImageFilter.BLUR)
display(img_blur)
```

```python
img_contour = img_50.filter(ImageFilter.CONTOUR)
display(img_contour)
```

```python
img_edged = img.filter(ImageFilter.FIND_EDGES)      # branches look beautiful, leave upper part
display(img_edged)
```

```python
img_detailed = img_50.filter(ImageFilter.DETAIL)
display(img_detailed)
```

```python
img_smooth = img_50.filter(ImageFilter.SMOOTH)
img_detailed_smooth = img_detailed.filter(ImageFilter.SMOOTH_MORE)
display(img_smooth)
display(img_detailed_smooth)
```
